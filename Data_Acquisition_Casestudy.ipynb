```python
import pandas as pd
```

Extract Data:

Load the CSV file using pandas.read_csv()

```python
data_csv = pd.read_csv("/content/population.csv")
data_csv
```

Load the Excel file using pandas.read_excel()

```python
data_xlsv = pd.read_excel("/content/gdp.xlsx")
data_xlsv
```

Parse the XML file using xml.etree.ElementTree

```python
data_xml = pd.read_xml("/content/literacy_rate.xml")
data_xml
```

Load the JSON file using Pythons json module

```python
data_json = pd.read_json("/content/internet_users.json")
data_json
```

Transform Data:

Standardize country names across all datasets

```python
data_xml = data_xml.rename(columns={'name':'country'})
data_xml
```

```python
data_xlsv = data_xlsv.rename(columns={'Country':'country'})
data_xlsv
```

Create new columns such as internet penetration rate (internet users / population  100)

```python
data_json['internet_penetration_rate'] = data_json['internet_users']/data_csv['population']*100
```

Integrate Data:

Merge the datasets based on the country name

```python
data = pd.merge(data_csv,data_xml, on='country', how='outer')
data = pd.merge(data, data_json, on='country', how='outer')
data = pd.merge(data , data_xlsv, on='country', how='outer')

display(data)
```

Handle unmatched or missing records

```python
data.isnull().sum()
```

Validate the data for consistency and completeness

```python
data.dtypes
```

```python
check_literacy = data[(data['literacy_rate'] < 0) | (data['literacy_rate'] > 100)]
check_literacy

check_population = data[data['population'] <= 0]
check_population

check_penetration = data[(data['internet_penetration_rate'] < 0) | (data['internet_penetration_rate'] > 100)]
check_penetration
```

Analyze Data:

Load into SQLite:

```python
import sqlite3
```

Create an SQLite database

```python
conn = sqlite3.connect("country.db")
cursor = conn.cursor()
```

Load the final merged dataset into a database table

```python
data.to_sql("final_data",conn,index=False)
```

```python
cursor.execute("select * from final_data")
print(cursor.fetchall())
```

Write and execute SQL queries to extract insights

a. Countries with highest internet penetration

```python
cursor.execute("""select country, internet_penetration_rate
from final_data
order by internet_penetration_rate desc
limit 5""")
print(cursor.fetchall())
```

b. Average literacy rate

```python
cursor.execute("""select avg(literacy_rate)
from final_data
""")
print(cursor.fetchall())
```

c. Correlation between literacy rate and internet penetration

```python
cursor.execute(""" SELECT
    (AVG(literacy_rate * internet_penetration_rate) - AVG(literacy_rate) * AVG(internet_penetration_rate)) /
    (SQRT((AVG(literacy_rate * literacy_rate) - AVG(literacy_rate)*AVG(literacy_rate)) *
          (AVG(internet_penetration_rate * internet_penetration_rate) - AVG(internet_penetration_rate)*AVG(internet_penetration_rate))))
    AS correlation
FROM final_data;
""")
print(cursor.fetchall())
```

d. GDP per capita analysis

```python
cursor.execute("""select country , Round(GDP / population,2)
from final_data;
""")
print(cursor.fetchall())
```

e. Which countries have a GDP per capita above $10,000?

```python
cursor.execute(""" select country,Round(GDP / population) as GDP_per_capita
from final_data
where GDP_per_capita > 10000;
""")
print(cursor.fetchall())
```

f. What is the total population covered in the dataset?

```python
cursor.execute("""select sum(population)
from final_data
""")
print(cursor.fetchall())
```

g. Which countries have the lowest literacy rates, and how does that impact internet
access?

```python
cursor.execute("""select country, literacy_rate
from final_data
order by literacy_rate asc;
""")
print(cursor.fetchmany(10))
```

h. What are the top 5 wealthiest countries by total GDP, and how does that compare
with population size?

```python
cursor.execute("""select country, (GDP * population) as total_gdp
from final_data
order by total_gdp desc
limit 5;
""")
print(cursor.fetchall())
```

i. Find countries where internet users exceed 70% of the population.

```python
cursor.execute("""select country, internet_penetration_rate
from final_data
where internet_penetration_rate > 70
""")
print(cursor.fetchall())
```

j. What is the average GDP per capita for countries with internet penetration above
50%?

```python
cursor.execute(""" select country,AVG(GDP)
from final_data
where internet_penetration_rate > 50
""")
print(cursor.fetchall())
```

k. How many countries have a literacy rate above 90%, and what is their average
internet penetration?

```python
cursor.execute("""select country, AVG(literacy_rate), AVG(internet_penetration_rate)
from final_data
where literacy_rate >90
""")
print(cursor.fetchall())
```
