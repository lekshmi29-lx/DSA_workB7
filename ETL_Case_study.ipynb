Convert time to datetime

```python
df.info()
```

```python
df['time']=pd.to_datetime(df['time'])
df.info()
```

Drop records with missing latitude, longitude, mag

```python
df = df.dropna(subset=['latitude', 'longitude', 'mag'])
df
```

```python
df.isnull().sum()
```

```python
#df.info()
```

Filter only earthquakes with magnitude >= 4.0

```python
df_filter = df[(df['type'] == 'earthquake')|(df['mag'] >=40)]
df_filter
```

Add a column day_of_week from time

```python
df['day_of_week'] = df['time'].dt.day_name()
print(df[['time','day_of_week']])
```

Create a column severity_level based on magnitude:
 <4.0: "Low",
 4.0 - 6.0: "Moderate",
 6.0+: "High".

```python
def severity(mag):
  if mag < 4.0:
    return 'Low'
  elif 4.0 <= mag < 6.0:
    return 'Moderate'
  else :
    return 'High'

df['severity_level'] = df['mag'].apply(severity)
```

```python
df
```

Count number of earthquakes per place

```python
count = df[df['type'] == 'earthquake']['place'].value_counts()
print(count)
```

Compute average magnitude and max depth per day

```python
earthquake_summary = df[df['type'] == 'earthquake'].groupby(df['time'].dt.date).agg(avg_mag=('mag', 'mean'),max_depth=('depth', 'max')).reset_index()
print(earthquake_summary[['time','avg_mag','max_depth']])
```

Save:
1. Cleaned dataset to cleaned_earthquakes.csv
2. Summary data to earthquake_summary.csv

```python
df.to_csv("cleaned_earthquakes.csv",index=False)
earthquake_summary.to_csv("earthquake_summary.csv",index=False)
```
