UNIVARIANT

```python
# Distribution of Profit

plt.figure(figsize=(8,5))
sns.histplot(data["Profit"],bins=40, color="green")
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()

# Most orders have small profits near zero,
# but there are some large losses and a few very high profits acting as outliers.
```

```python
# Distribution of Sales

plt.figure(figsize=(8,6))
sns.histplot(data["Sales"], bins=50, kde=True, color="skyblue")
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

# Most sales are small,
# but a few very large orders create outliers and strongly affect total sales.
```

```python
# Count of Orders by Category

plt.figure(figsize=(6,5))
sns.countplot(x="Category", data=data, hue="Category", palette="Set2", legend=False)
plt.title("Count of Orders by Category")
plt.show()

# The 'Office Supplies' category has the highest number of orders,
# followed by 'Furniture' and then 'Technology'.
# This suggests that customers purchase office supplies more frequently,
# whereas technology products, though fewer in order count, may contribute
# higher sales per order.
```

BIVARIANT

```python
# Sales by Category
sales_by_category = data.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,6))
sales_by_category.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()

# From the bar chart, we can see that the 'Technology' category generates the highest total sales,
# followed by 'Furniture', while 'Office Supplies' has the lowest total sales among the three.
```

```python
# Sales vs Profit (numeric vs numeric)

plt.figure(figsize=(8,6))
sns.scatterplot(x="Sales", y="Profit", hue="Category", data=data, alpha=0.6)
plt.title("Sales vs Profit by Category")
plt.show()

# The scatter plot shows that higher sales usually lead to higher profits,
# but many orders still result in losses, especially in Furniture and Office Supplies.
# Technology often gives higher profits on large sales, but it also has some big losses.
# Overall, profit does not always increase with sales because discounts and costs vary across categories.
```

```python
# Average Sales by Region

plt.figure(figsize=(6,5))
sns.barplot(x="Region", y="Sales", data=data, hue="Region",estimator=sum, errorbar=None, palette="muted",legend=False)
plt.title("Total Sales by Region")
plt.show()

# The West region has the highest total sales, followed by the East and Central regions.
# The South region shows the lowest sales performance overall.
# This suggests that business strategies could be focused more on improving sales in the South region.
```

```python
#Profit vs Sales scatter plot

plt.figure(figsize=(8,6))
plt.scatter(data["Sales"], data["Profit"], alpha=0.5, c="green")
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()

# Sales and Profit generally move in the same direction, but not always.
# Some high sales still result in losses (negative profit).
# This means discounts, product type, or costs may reduce profitability.
```

MULTIVARIANT

```python
# The relationship between Sales and Profit varies across both Category and Region.

plt.figure(figsize=(10,6))
sns.scatterplot(x="Sales", y="Profit", hue="Category", style="Region", data=data, alpha=0.7)
plt.title("Sales vs Profit by Category and Region")
plt.show()

# Technology products usually give higher profits with large sales but sometimes face big losses.
# Furniture and Office Supplies often have more losses, especially in the Central and South regions.
# Profitability depends on both product type and regional market conditions like pricing, demand, and discounts.
```

```python
# Example: Heatmap of correlation between numerical columns

plt.figure(figsize=(8,6))
corr = data[["Sales", "Profit", "Quantity", "Discount"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()

# Sales and Profit have a moderate positive correlation (0.48),higher sales generally lead to higher profits.
# Profit and Discount show a negative correlation (-0.22), increasing discounts reduces profit.
# Sales and Quantity have a weak positive correlation (0.2), higher quantity slightly contributes to higher sales.
# Discount has very little relationship with Sales (-0.028) and Quantity (0.0086), but it clearly harms Profit.
```

```python
# Shows how Sales, Profit, Quantity, Discount interact with each other.

plt.figure(figsize=(10,6))
sns.pairplot(data[["Sales", "Profit", "Quantity", "Discount"]], diag_kind="kde")
plt.suptitle("Pairplot of Sales, Profit, Quantity, and Discount", y=1.02)
plt.show()

# Sales vs Profit : positively related, but some orders with high sales still result in losses, likely due to discounts or pricing.
# Quantity vs Profit : do not have a clear linear relationship; selling more units does not always increase profit.
# Discount vs Profit : negatively correlated, meaning higher discounts often reduce profit.
# Sales & Profit : Highly skewed, a few extreme outliers.
# Quantity : Mostly small orders (5–10 units).
# Discount : Not evenly distributed, applied selectively.
```

```python
#Boxplot (Profit Distribution by Category & Region)
plt.figure(figsize=(10,6))
sns.boxplot(x="Category", y="Profit", hue="Region", data=data)
plt.title("Profit Distribution by Category and Region")
plt.show()

# Furniture : Large variation, many losses (negative profits).
# Office Supplies : Lower variation, but has extreme outliers.
# Technology : Generally higher profits, but still some losses.
# Across Regions : Medians are similar, but all have strong outliers.
```
