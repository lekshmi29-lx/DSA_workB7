Handling Missing Values

```python
# Identify missing values in each column.

house_price.isna().sum()
```

```python
house_price = house_price.dropna(subset=['Sale Price'])
```

```python
from scipy.stats import skew
```

```python
skewness = house_price['No of Bathrooms'].skew
skewness()
```

```python
house_price.loc[:,'No of Bathrooms']  = house_price['No of Bathrooms'].fillna(house_price['No of Bathrooms'].median())
```

```python
house_price.loc[:,'Flat Area (in Sqft)'] = house_price['Flat Area (in Sqft)'].fillna(house_price['Area of the House from Basement (in Sqft)']+house_price['Basement Area (in Sqft)'])
```

```python
house_price.loc[:,'Lot Area (in Sqft)'] = house_price['Lot Area (in Sqft)'].fillna(house_price['Lot Area after Renovation (in Sqft)'])
```

```python
house_price = house_price.dropna(subset=['Zipcode','Latitude','Longitude'])
```

```python
house_price = house_price.drop('No of Times Visited',axis = 1)
```

```python
house_price = house_price.drop(['Latitude','Longitude'],axis = 1)
```

```python
house_price = house_price.dropna(subset=['Living Area after Renovation (in Sqft)'])
```

```python
house_price.loc[:,'Area of the House from Basement (in Sqft)'] = house_price['Area of the House from Basement (in Sqft)'].fillna(house_price['Flat Area (in Sqft)']-house_price['Basement Area (in Sqft)'])
```

```python
house_price = house_price.drop(['ID','Area of the House from Basement (in Sqft)','Basement Area (in Sqft)'],axis=1)
```

```python
house_price = house_price.drop(['Lot Area (in Sqft)','Renovated Year'],axis=1)
```

```python
house_price = house_price.drop(['Flat Area (in Sqft)','Date House was Sold'],axis = 1)
```

Handle Outliers

```python
# plot all numerical_columns using Boxplot to Identify the Outlier

numerical_col = house_price.select_dtypes(include = ['float64','int64'])
for i ,col in enumerate(numerical_col):
  plt.figure(figsize=(8,8))
  sns.boxplot(data=numerical_col,y=col)
  plt.title(col)
  plt.show()
```

```python
# Handle Outlier - Sles Price
# Original Sale Price before handle outlier

plt.subplot(1,2,1)
sns.histplot(house_price['Sale Price'], bins=30, kde=True)
plt.title("Original Sale Price")
```

```python
# Log Transform in target
house_price['Sale Price'] = np.log(house_price['Sale Price'])
```

```python
# After using log Transform - sales price

plt.subplot(1,2,2)
sns.histplot(house_price['Sale Price'], bins=30, kde=True)
plt.title("Log-Transformed Sale Price")
```

```python
# Handle outlier - Number of Bedrooms
q1 = np.quantile(house_price['No of Bedrooms'],0.25)
q3 = np.quantile(house_price['No of Bedrooms'],0.75)
iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
bedroom_outliers = []
for i in house_price['No of Bedrooms']:
  if i > upper_bound or i < lower_bound:
    bedroom_outliers.append(i)
bedroom_outliers
```

```python
# After handle the outlier by using iqr method - No of Bedrooms

house_price['No of Bedrooms'] = house_price['No of Bedrooms'].clip(lower = lower_bound,upper= upper_bound)
plt.boxplot(house_price['No of Bedrooms'])
```

```python
# Handle outlier - No of Bathrooms

q1 = np.quantile(house_price['No of Bathrooms'],0.25)
q3 = np.quantile(house_price['No of Bathrooms'],0.75)
iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
bathroom_outliers = []
for i in house_price['No of Bathrooms']:
  if i > upper_bound or i < lower_bound:
    bathroom_outliers.append(i)
bathroom_outliers
```

```python
# After handle the outlier by using iqr method - No of Bathrooms

house_price['No of Bathrooms'] = house_price['No of Bathrooms'].clip(lower = lower_bound,upper= upper_bound)
plt.boxplot(house_price['No of Bathrooms'])
```

```python
# Handle Outlier - Overall Grade

plt.subplot(1,2,1)
sns.histplot(house_price['Overall Grade'], bins=30, kde=True)
plt.title("Original Overall Grade")
```

```python
q1 = np.quantile(house_price['Overall Grade'],0.25)
q3 = np.quantile(house_price['Overall Grade'],0.75)
iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
grade_outliers = []
for i in house_price['Overall Grade']:
  if i > upper_bound or i < lower_bound:
    grade_outliers.append(i)
grade_outliers
```

```python
# After handle the outlier by using iqr method - No of Bathrooms

house_price['Overall Grade'] = house_price['Overall Grade'].clip(lower = lower_bound,upper= upper_bound)
plt.boxplot(house_price['Overall Grade'])
```

```python
# Handle Outlier - Living Area after Renovation (in Sqft)

plt.subplot(1,2,1)
sns.histplot(house_price['Living Area after Renovation (in Sqft)'], bins=30, kde=True)
plt.title("Original Living Area after Renovation (in Sqft)")
```

```python
from scipy.stats.mstats import winsorize
```

```python
# use winsorizing method

house_price['Living Area after Renovation (in Sqft)'] = winsorize(house_price['Living Area after Renovation (in Sqft)'],limits=(0.05,0.05))
```

```python
plt.boxplot(house_price['Living Area after Renovation (in Sqft)'])
```

```python
# Handle Outlier - Lot Area after Renovation (in Sqft)

plt.subplot(1,2,1)
sns.histplot(house_price['Lot Area after Renovation (in Sqft)'], bins=30, kde=True)
plt.title("Original Lot Area after Renovation (in Sqft)")
```

```python
# using winsorized method

house_price['Lot Area after Renovation (in Sqft)'] = winsorize(house_price['Lot Area after Renovation (in Sqft)'],limits=(0.05,0.15))
```

```python
plt.boxplot(house_price['Lot Area after Renovation (in Sqft)'])
```
